```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
region=pd.read_csv('kiva_mpi_region_locations.csv')
```

```python
region.head()
```

```python
region.info()
```

```python
region_copy=region.dropna()
```

```python
region_copy.info()
```

```python
mpi_country=region_copy.groupby('country')['MPI'].mean()
mpi_country
```

```python
mpi_country=pd.DataFrame(mpi_country)
```

```python
mpi_country.reset_index()
```

```python
mpi_country=mpi_country.sort_values(by=['MPI'])
```

```python
mpi_country
```

```python
mpi_region=region_copy.groupby('world_region')['MPI'].mean()
mpi_region
```

```python
mpi_region=pd.DataFrame(mpi_region)
```

```python
mpi_region.reset_index()
```

```python
mpi_region=mpi_region.sort_values(by=['MPI'])
```

```python
mpi_region.plot(kind='bar')
ax  = plt.gca()
plt.title('MPI by world region')
for i in range(6):
    ax.text(x=i,y=mpi_region['MPI'][i],s=round(mpi_region['MPI'][i],2))
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Store latitude and longitude in locations
locations = region_copy[["lat", "lon"]]

# Fill NaN values and convert to float
MPI = region_copy["MPI"].astype(float)
```

```python
import gmaps

# Google developer API key
from config import gkey

# Access maps with unique API key
gmaps.configure(api_key=gkey)
```

```python
# Created a heatmap that shows the poverty level (MPI) round the world 
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=MPI, 
                                 dissipating=False, max_intensity=1,
                                 point_radius=3)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
