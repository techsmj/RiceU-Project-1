```python
import pandas as pd
from apiExchange import api_key
import requests
import json
```

```python
CPI_Data = pd.read_csv("CorruptionPerceptionIndex.csv")
CurrID = pd.read_csv("CurrencyCodes.csv")
CPI_Data.head()
```

```python
CurrID.head()
```

```python
CPI_CurrID = CPI_Data.merge(CurrID, on="Country")
CPI_CurrID.rename(columns={'Code': 'Currency ID'}, inplace=True)
CPI_CurrID.head()
```

```python
base_curr = "USD"
base_url = "https://openexchangerates.org/api/latest.json?app_id="
```

```python
query_url = base_url + api_key + "&base=" + base_curr
```

```python
rates_response = requests.get(query_url)
rates_json = rates_response.json()
rates_json
```

```python
sub_df = rates_json.pop('rates')
sub_df
```

```python
sub_dfRATES = pd.DataFrame.from_dict(sub_df, orient='index')
sub_dfRATES
```

```python
ExchangeRatesDF= pd.DataFrame(sub_dfRATES).reset_index()
ExchangeRatesDF.rename(columns={'index': 'Currency ID', 0:'Rate in USD'}, inplace=True)
ExchangeRatesDF
```

```python
CPI_CurrID = CPI_CurrID.merge(ExchangeRatesDF, on="Currency ID")
CPI_CurrID
```
