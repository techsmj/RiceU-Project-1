```python
import pandas as pd
import numpy as np
import scipy as stats
```

```python
#Upload all csv files

loan_data = pd.read_csv('loan_clean.csv')
cpi_exc_data = pd.read_csv('CPI_and_Exchange.csv')
continent = pd.read_csv('country-and-continent-codes-list-csv.csv')
loan_types = pd.read_csv('loan_theme_ids.csv')
mpi = pd.read_csv('kiva_mpi_region_locations.csv')

```

```python
#to mergedata the columns need to match at 'country'
loan_data.rename(columns={'country':'Country'}, inplace=True)
loan_data.head()
```

```python
#merge and #preview merged data
loan_data=loan_data.merge(cpi_exc_data, on='Country')
loan_data.head()
```

```python
#reset and rename columns to match above 'country code' column
df= pd.DataFrame(continent).reset_index()
df.rename(columns={'Two_Letter_Country_Code': 'country_code'}, inplace=True)
df.head()
```

```python
#merge on country code

loan_data=loan_data.merge(df, on='country_code')
loan_data.head()
```

```python
#add loan types to our data on 'id' and reset dataframe
loan_data=loan_data.merge(loan_types, on='id')
pd.DataFrame(loan_data).reset_index()
loan_data.head()
```

```python
#add in mpi data
mpi_update= pd.DataFrame(mpi).reset_index()
mpi_update.rename(columns={'country': 'Country'}, inplace=True)
mpi_update.head()
```

```python
#merge mpi with loan data
loan_data=loan_data.merge(mpi_update, on='Country')
loan_data.head()
```

```python
#drop unnecessary columns
loan_data_update = loan_data.drop(columns=['Unnamed: 0','funded_amount','region_x','term_in_months','postedtime','disbursedtime','ISO3','Region','CPI score 2018','Rank 2018', 'Sources 2018', 'Standard error 2018','Rank 2017', 'Sources 2017', 'Standard error 2017', 'CPI score 2016','Sources 2016', 'Standard error 2016', 'CPI score 2015', 'Sources 2015','Standard error 2015', 'Sources 2014','Standard error 2014', 'CPI Score 2013', 'Sources 2013','Standard error 2013', 'CPI Score 2012', 'Sources 2012','Standard error 2012', 'Capital', 'Currency Name','Rate in USD_y', 'index_x', 'Country_Name', 'Three_Letter_Country_Code','Country_Number','Loan Theme ID','Partner ID','index_y','LocationName','ISO','region_y','geo'], axis=1,inplace=True)
```

```python
#save to csv
loan_data.to_csv('MasterDataSet_updated.csv')
```
